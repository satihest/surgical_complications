## Make list of all fam-IDs and split 90/10

```python
list_fam = pd.read_csv(r'E:/Data/UkBiobank/ukb_10.fam', delimiter = ' ', header = None)
```

```python
list_fam['Yes'] = 1
list_fam
```

# Make split like this: All general + 80% surgical. Reserve 20% of surgical for test. 

## Afib

```python
df_nonsurg = pd.read_csv(r'E:/Mathias/phenotypes/nosurg_pheno_afib.txt', delimiter = '\t', header = None)
df_nonsurg
```

```python
df_surg = pd.read_csv(r'E:/Mathias/phenotypes/surg_pheno_afib.txt', delimiter = '\t', header = None)
df_surg
```

split surgical cohort

```python
training_data_surg = df_surg.sample(frac=0.8, random_state=25)
testing_data_surg = df_surg.drop(training_data_surg.index)

print(f"No. of training examples: {training_data_surg.shape[0]}")
print(f"No. of testing examples: {testing_data_surg.shape[0]}")
```

Append non-surgical cohort with training surgical cohort

```python
frames = [df_nonsurg, training_data_surg]
df_training = pd.concat(frames)
df_training = list_fam.merge(df_training, how='right', on= 0)
df_training = df_training[df_training['Yes'] == 1]
df_training.rename(columns = {0: 'ID'}, inplace=True)
df_training = df_training[['ID']]
df_training
```

Above is IDs for training + validation set (all general + 80% surgical). Below create IDs of test set (20% surgical)

```python
df_test = testing_data_surg
df_test.rename(columns = {0: 'ID'}, inplace=True)
df_test = df_test[['ID']]
df_test
```

Check for overlap in cohorts and then export plink binary files for use in PP

```python
overlap = pd.merge(df_test, df_training, on='ID', how='inner')

if overlap.empty:
    print("No overlap between dfs")
else:
    print("There is some overlap. Recheck")
    print(overlap)
```

```python
df_test.to_csv(r'E:\Mathias\phenotypes\test_ids_afib_v2.txt', header=False, index=None)
df_training.to_csv(r'E:\Mathias\phenotypes\training_ids_afib_v2.txt', header=False, index=None)
```

```python
training_data = list_fam.sample(frac=0.9, random_state=25)
testing_data = list_fam.drop(training_data.index)

print(f"No. of training examples: {training_data.shape[0]}")
print(f"No. of testing examples: {testing_data.shape[0]}")
```

Make phenotype/target data for the test and training sets for afib, vte and pneumo

```python
all_afib = pd.read_csv(r'E:/Mathias/phenotypes/afib_all_pheno.txt', delimiter = '\t', header = None)
all_afib.rename(columns = {0: 'ID', 2: 'afib'}, inplace=True)
all_afib = all_afib[['ID', 'afib']]
afib_train_v2 = df_training.merge(all_afib, how='left', on= 'ID')
afib_train_v2
```

```python
afib_test_v2 = df_test.merge(all_afib, how='left', on= 'ID')
afib_test_v2
```

```python
afib_train_v2.to_csv(r'E:\Mathias\phenotypes\afib_train_v2.txt', header=True, index=None)
afib_test_v2.to_csv(r'E:\Mathias\phenotypes\afib_test_v2.txt', header=True, index=None)
```
